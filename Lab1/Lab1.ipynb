***Lab1 Data Mining***
---------------------------------

- data contains 150 vector
- each vector represents a sample of the Iris data
- there are 3 classes of flowers, each class have 50 samples
- each vector is a tuple of 4 values of attributes or features(petal length and width , sepal length and width)

```python
from sklearn import datasets
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
df = pd.DataFrame(data= np.c_[iris.data, iris.target],
                     columns= np.append(iris.feature_names, 'class'))
```

```python
df
```

```python
X = iris.data
y = iris.target
```

```python
X.shape
```

```python
y.shape
```

**Visualization 1**
- we can notice that values distribution of each feature of last 2 classes looks approximatily the same
- petal length in class 2,3 values are higher
- sepal width in class 1 is higher

```python
f, axs = plt.subplots(iris.target_names.shape[0], 1, figsize=(15,15))
for i in range(0,iris.target_names.shape[0]):
    axs[i].plot(iris.feature_names,df[df['class'] == i].drop(['class'], axis=1).T,'ro')
    axs[i].set_title('Class '+ iris.target_names[i])
```

**Visualization 2**
- in most of the plots we can notice that most of the data lies near mean value
- ranges of values of corresponding attributes of each class varies

```python
for i in range (0,iris.target_names.shape[0]):
    df[df['class'] == i].drop(['class'], axis=1).hist(edgecolor='black', linewidth=1.2)
    fig=plt.gcf()
    fig.set_size_inches(15,8)
    fig.suptitle('Class Name: '+iris.target_names[i], fontsize=20)
    plt.show()
```

**Visualization 3**
- sepal length and petal length are positivly correlated
- petal width and sepal length are positivly correlated
- petal width and petal length are strongly positivly correlated

```python
pd.scatter_matrix(df.drop(['class'], axis=1), alpha = 1, figsize = (14,8), diagonal = 'kde');
```

```python
import seaborn as sns; sns.set()
correlation = df.drop(['class'], axis=1).corr(method='pearson')
ax = sns.heatmap(correlation,annot=True)
```
